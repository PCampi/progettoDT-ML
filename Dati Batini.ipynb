# Analisi completezza dati per continente

```python
import pandas as pd
```

```python
cc = pd.read_csv('./Dataset/finali/Country-continent.csv')

co2 = pd.read_csv('./Dataset/Batini/CO2.csv')
```

```python
co2 = co2.rename(columns={'Country Code': 'alpha-3'})
co2 = co2.drop(columns=['Series Name', 'Series Code'])
co2.head()
```

```python
cc = cc.drop(columns=['alpha-2', 'country-code', 'iso_3166-2', 'region-code', 'sub-region-code'])
```

### Join dei dataset sul codice alpha-3

```python
co2_joined = pd.merge(co2, cc.drop(columns=['name']), on='alpha-3')
co2_joined.head()
```

```python
co2_gr = co2_joined.groupby(by='region')
co2_gr.groups
```

```python
# Funzioni per analisi completezza
def completezza_attributi(dataset):
    """Calcola la completezza degli attributi di un dataset."""
    totale = dataset.shape[0]
    non_nulli = dataset.count()
    
    return non_nulli, totale

def null_in_tuple(dataset):
    """Calcola la completezza di tupla di un dataset."""
    num_attributi = dataset.shape[1]
    num_null_in_tupla = [dataset.iloc[i, :].isna().sum() for i in range(len(dataset))]
    
    return num_null_in_tupla

def completezza_tabella(dataset):
    """Calcola la completezza di tabella in un dataset."""
    totali_non_null = dataset.count().sum()
    totali = dataset.shape[0] * dataset.shape[1]
    
    return totali_non_null, totali
```
